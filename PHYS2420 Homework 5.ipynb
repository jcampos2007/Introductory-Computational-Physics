```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
m1 = 8.52e-3 #kg
m4 = 3.408e-2 #kg
rf = 5.8e-2 #m
dy = 2.03 #m
g = 9.8 #m/s^2
rho = 1.045 #kg/m^3
A = np.pi * rf * rf
```

```python
def derivatives_quadratic(t, S, m, C):
  y = S[0]
  v = S[1]
  F = (m * g) - (0.5 * rho * A * C * v * v)
  return [v, F/m]

def derivatives_linear(t, S, m, b):
  y = S[0]
  v = S[1]
  F = (m * g) - (b * v)
  return [v, F/m]

def event_hit_ground(t, S, arg1, arg2):
  return S[0] + dy

event_hit_ground.terminal = True
event_hit_ground.direction = 1

def simulate_fall_quadratic(m, C):
  S_initial = [0,0]
  t_span = (0,5)

  sol = solve_ivp(derivatives_quadratic, t_span, S_initial, events = event_hit_ground, args = (m, C), method = 'RK45')
  return sol.t_events[0][0]

def simulate_fall_linear(m, b):
  S_initial = [0,0]
  t_span = (0,5)

  sol = solve_ivp(derivatives_quadratic, t_span, S_initial, events = event_hit_ground, args = (m, b), method = 'RK45')
  return sol.t_events[0][0]
```

```python
print("Quadratic model, no air resistance, m1:", simulate_fall_quadratic(m1, 0))
print("Linear model, no air resistance, m1:", simulate_fall_linear(m1, 0))
print("Analytic solution, no air resistance, m1:", np.sqrt(2 * dy / g))
```
