```python
#Problem 8

import numpy as np

realSinh = np.sinh(2)
ourSinh = 0.5 * (np.exp(2) - np.exp(-2))
relativeError = (ourSinh - realSinh) / realSinh

print("sinh(2) = ", realSinh)
print("Our expression for sinh(2) = ", ourSinh)
print("Relative error: ", relativeError)
```
