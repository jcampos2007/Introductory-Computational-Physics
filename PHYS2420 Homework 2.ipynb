```python
# Problem 4, dt = 0.01

import numpy as np

g = 9.81 #m/s

p0 = np.array([-0.02,-0.01,-0.02]) #kg m/s
x0 = np.array([-0.2,-0.61,0]) #m
m = 0.011 #kg
k = 0.930 #N/m
l = 0.250 #m
dt = 0.01 #s
tTotal = 0.1 #s

x = x0.copy()
p = p0.copy()

timesteps = int(tTotal / dt)
for i in range(timesteps):

  xMag = np.linalg.norm(x)
  if (xMag < l):
    fTension = np.array([0,0,0])
  else:
    fTensionMag = k*(xMag - l)
    fTension = -x*(fTensionMag / xMag)
  fGravity = np.array([0,-m*g, 0])
  fNet = fTension + fGravity

  dp = fNet * dt
  p += dp

  v = p / m
  dx = v * dt
  x += dx

print("Position after 0.1 seconds: ", x, " m")
```

```python
# Problem 6

def calculate_gravity(m1, m2, r):
  return (6.674E-11 * m1 * m2 * (r ** -2))

mEarth = 5.97E24 #kg
mMoon = 7.34E22 #kg
rAvgEarthMoon = 3.84E8 #m

gravity = calculate_gravity(mEarth, mMoon, rAvgEarthMoon)
print("Average gravitational force between Earth and Moon: ", gravity, " N")
```

```python
# Problem 7

import numpy as np
import matplotlib.pyplot as plt

times = np.linspace(0,5,51)
y = np.empty(51)

i = 0
for t in times:
  y[i] = (25 * t) - (4.9 * t * t)
  i+=1

plt.plot(times, y)
plt.xlabel("Time (s)")
plt.ylabel("Height (m)")
plt.show()
```
